```python
import glob
import os

files = glob.glob("/mnt/HDD10TB-2/ohtani/dataset_info/filesize*.txt")
for file in files:
    df = pd.read_csv(file, names=['filesize'])
    ax = df.plot.hist(bins=100)
    ax.set_xlabel('filesize');
    ax.set_xscale('log')
    ax.set_xlim(10**3-100,10**7+1000)    #X軸の範囲を設定
    ax.set_ylim(0, 2000)
    file_a = os.path.split(file)[1]
    q = file_a.split('.')[0].split('_')[1]
    print(q)
    parameters = {'xtick.labelsize': 18,
          'ytick.labelsize': 18,
          'figure.figsize': (8,6)}
    plt.rcParams.update(parameters) 
#     plt.figure(figsize=(8,6))
    fig_save = 'plt.savefig("filesize_fig/filesize_' + str(q) + '.pdf")'
    exec(fig_save)
    plt.show()
```
