```python
df = pd.read_csv('../../dataset_info/stream_info_val.txt',
                 names=['width', 'height', 'frame_rate', 'bit_rate'],
                 delimiter=','
                )

bit_rate = df['bit_rate'].values
bit_rate_avg = np.average(bit_rate) / 1000
print(len(bit_rate))
print(bit_rate_avg)
```

```python
ax = df.bit_rate.plot.hist(bins=200)
ax.set_xlabel('bit rate');
ax.set_xscale('log')
ax.set_xlim(10**3,10**7) 
parameters = {'xtick.labelsize': 11,
          'ytick.labelsize': 11}
plt.rcParams.update(parameters)   #X軸の範囲を設定
```

```python
data = df['bit_rate'].values
print(np.mean(data))
plt.boxplot(data, showfliers=False, showmeans=True)
```

```python
import glob
import os

files = glob.glob("/mnt/HDD10TB-2/ohtani/dataset_info/stream_info*.txt")
for file in files:
    df = pd.read_csv(file, names=['width', 'height', 'frame_rate', 'bit_rate'], delimiter=',')
    ax = df.bit_rate.plot.hist(bins=200)
    ax.set_xlabel('bit rate');
    ax.set_xscale('log')
    ax.set_xlim(10**3-100,10**7+100)    #X軸の範囲を設定
    ax.set_ylim(0, 1000)
    file_a = os.path.split(file)[1]
    q = file_a.split('.')[0].split('_')[2]
    print(q)
    parameters = {'xtick.labelsize': 18,
    'ytick.labelsize': 18,
    'figure.figsize': (8,6)}
    plt.rcParams.update(parameters) 
    fig_save = 'plt.savefig("stream_fig/bit_rate_' + str(q) + '.pdf")'
    exec(fig_save)
    plt.show()
```
