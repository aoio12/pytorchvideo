```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
%matplotlib inline
```

```python
df = pd.read_csv('../../dataset_info/stream_info_val.txt',
                 names=['width', 'height', 'frame_rate', 'bit_rate'],
                 delimiter=','
                )
ax = df.bit_rate.plot.hist(bins=200)
ax.set_xlabel('bit rate [b/s]');
ax.set_xscale('log')
ax.set_xlim([10**3-100,10**8])    #X軸の範囲を設定
ax.set_ylim(0, 800)

parameters = {'xtick.labelsize': 16,
'ytick.labelsize': 16,
'axes.labelsize': 16,
'axes.titlesize': 16,
'figure.figsize': (8,6),}
plt.rcParams.update(parameters)
plt.savefig("stream_fig/original_bit_rate.pdf")
plt.show()
```
