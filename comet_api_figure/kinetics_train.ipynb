## train top1 accuracy

```python
fig, ax = plt.subplots()
ax.set_xlabel('epoch')
ax.set_ylabel('accuracy')
ax.set_title('train top1 accuracy')
ax.grid()
# ax.set_xlim(0.9, 5.1)
ax.set_ylim(59, 66)
labels = list(range(1, 6))
colors = ["red", "green", "blue"]
original_color = "black"
train_crfs = [20, 30, 40]
# train_crfs = [20, 30]

#original
train_top1_list = []
query = ((model_name == 'x3d_m') & (train_crf_param == -10) & (train_gop_param == -10) & tagged)
experiments = api.query('ohtani', 'ffmpeg', query)
train_top1 = experiments[0].get_metrics('epoch_train_top1')
for i in range(len(train_top1)):
    train_top1_score = float(train_top1[i].get('metricValue'))
    train_top1_list.append(train_top1_score)
ax.plot(labels, train_top1_list, '-o', color=original_color, label='original')

#圧縮動画
for train_crf in train_crfs:
    train_top1_list = []
    query = ((model_name == 'x3d_m') & (train_crf_param == train_crf) & (train_gop_param == 12) & tagged)
    experiments = api.query('ohtani', 'ffmpeg', query)
    train_top1 = experiments[0].get_metrics('epoch_train_top1')
    for i in range(len(train_top1)):
        train_top1_score = float(train_top1[i].get('metricValue'))
        train_top1_list.append(train_top1_score)
    
    index = train_crfs.index(train_crf)
    ax.plot(labels, train_top1_list, '-o', color=colors[index], label=train_crf)

parameters = {'xtick.labelsize': 16,
        'ytick.labelsize': 16,
        'axes.labelsize': 16,
        'axes.titlesize': 16,
        }
plt.rcParams.update(parameters) 
ax.legend(loc=0) 
fig.tight_layout()  # レイアウトの設定
plt.savefig('../acc_fig/ffmpeg_train/kinetics400/x3d_m/train_acc_top1.pdf')
plt.show()



```
